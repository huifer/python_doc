# Matplotlib figure
## matplotlib.pyplot.figure
```
matplotlib.pyplot.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, FigureClass=<class 'matplotlib.figure.Figure'>, clear=False, **kwargs)
```

- num
    - 画布的唯一标识,不填自增
- figsize
    - 画布大小,单位英寸
- dpi
    - 分辨率
- facecolor
    - 背景颜色
- edgecolor
    - 边框颜色
- frameon 
    - 是否绘制图框,默认值True

- clear
    - 是否清除相同文字,默认值False
    
## 默认值

````python
import matplotlib
print(matplotlib.rcParams['figure.figsize'])
print(matplotlib.rcParams['figure.dpi'])
print(matplotlib.rcParams["figure.facecolor"])
print(matplotlib.rcParams["figure.edgecolor"])    
````

````python
import matplotlib.pyplot as plt
import itertools
import warnings


fig = plt.figure(
    num=1,figsize=(10,10),
    dpi=300,facecolor="red",edgecolor="blue",clear=True,linewidth=4)
axes1=plt.axes((0,0,0.3,0.3))
line=plt.Line2D([0,1],[0,1])
l=axes1.add_artist(line)
fig.add_axes(axes1)
fig.show(warn=False)
````
