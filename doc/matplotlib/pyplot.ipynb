# pyplot
## 简述
- pyplot 作为 matplotlib 绘图方面的核心,Title、Y axis label、X axis label、 Major tick、Legend等都是在这个基础上进行追加
## 导入
`import matplotlib.pyplot as plt`

````python
import matplotlib.pyplot as plt
````

## 绘图
> 官方文档:https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
- 函数
```
matplotlib.pyplot.plot(*args, **kwargs)
```
### 例子
- plot(x, y)  根据 x,y 绘制图像,颜色图例均为默认

````python
plt.plot([1,2,3,4],[1,2,3,4])
plt.show()
````

- plot(x, y, 'bo')  根据 x,y 绘制蓝色圆圈图像

````python
plt.plot([1,2,3,4],[1,2,3,4],'bo')
plt.show()
````

- plot(y)         根据 y 进行绘制 x至为 (max(y))-1

````python
plt.plot([1,2,3,4])
plt.show()
````

- plot(y, 'r+')    绘制 (max(y))-1 ,y 进行绘制, 颜色红,图例+

````python
plt.plot([1,2,3,4],'r+')
plt.show()
````

### 坐标轴范围
- 函数
    - `matplotlib.pyplot.axis`
    - 传递的参数
        - [xmin，xmax，ymin，ymax]

````python
plt.plot([1,2,3,4],[1,2,3,4],'bo')
plt.axis([0,5,0,6])
plt.show()
````

### 多个图像
- 在`matplotlib.pyplot.plot`传递参数(x1,y1,图例1,x2,y2,图例2)

````python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
````

### 线条属性(样式)
- `matplotlib.lines.Line2D` 核心类
    - 文档: https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D
#### 常用属性
- linewidth 线宽

````python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)
line , = plt.plot(t, t, 'r--', linewidth=4)
plt.show()
````

## 多图, 多轴 figure
详细文档可看: [figure](figure.ipynb)

````python
import matplotlib.pyplot as plt
plt.figure(1)                       # 第一个图形
plt.subplot(211)                    # 第一个图形的第一个子图
plt.plot([1, 2, 3])
plt.subplot(212)                    # 第一个图形的第二个子图
plt.plot([4, 5, 6]) 


plt.figure(2)                       # 第二个图形
plt.plot([4, 5, 6])                 # 默认创建 subplot(111)

plt.figure(1)                       # 当前是图形 1，subplot(212)
plt.subplot(211)                    # 将第一个图形的 subplot(211) 设为当前子图
plt.title('Easy as 1, 2, 3')        # 小标题
plt.suptitle("Categorical Plotting") # 大标题
plt.show()
````

## 文本操作
- 详细文档可看: [Text](Text.ipynb)
- 官方文档: https://matplotlib.org/tutorials/text/text_intro.html
- 一般的使用  `xlabel()`, `ylabel()` , `title()` 设置 x 轴标题, 设置 y 轴标题,设置标题

````python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()







````
